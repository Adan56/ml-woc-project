```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Desktop/binary_classification_train.csv')
data.loc[0:40]
```

```python
data = np.array(data)
X = data[:,1:21]
Y = data[:,21]
m,n = X.shape
```

```python
def sigmoid_fn(z):
    sgmd = 1/(1 + np.exp(-z))
    return sgmd
```

```python
def gradient_function(X, Y, w, b): 
    
    z = np.dot(X,w) + b
    f_wb_x = sigmoid_fn(z)
    dj_dw = np.dot(f_wb_x - Y,X) / m
    dj_db = np.sum(f_wb_x - Y) / m
    
    return dj_db, dj_dw
```

```python
def cost_function(X,Y,w,b):
    z = np.dot(X,w) + b
    f_wb_x = sigmoid_fn(z)
    cost = np.dot(Y,np.log(f_wb_x)) + np.dot(1-Y,np.log(1-f_wb_x))
    cost = cost*(-1/m)
    return cost
```

```python
def gradient_descent(X, Y, w_in, b_in, learning_rate, iterations): 
    
    w = w_in
    b = b_in
    
    for i in range(iterations):
        
        dj_db,dj_dw = gradient_function(X, Y, w, b)  
        w = w - learning_rate * dj_dw               
        b = b - learning_rate * dj_db

        if i%500 == 0:
            cost = cost_function(X,Y,w,b)
            print(f"iterations = {i}, cost = {cost}")
        
    return w, b
```

```python
initial_w = np.zeros(n)
initial_b = 0.
iterations = 5000
learning_rate = 0.00003

w_final, b_final = gradient_descent(X, Y, initial_w, initial_b, learning_rate, iterations)
```

```python
z = np.dot(X,w_final) + b_final
f_wb_x = sigmoid_fn(z)

print(f"b,w found by gradient descent: {b_final},{w_final} ")

results = []
Y_pred =[]
for i in range(m):
    # if(f_wb_x[i] >= 0.5):
    #     print(f"prediction : 1 , target value: {Y[i]}")
    # else:
    #     print(f"prediction : 0 , target value: {Y[i]}")
    prediction = 1 if f_wb_x[i] >= 0.5 else 0
    actual = Y[i]
    is_prediction_correct = "Correct Prediction" if prediction==actual else "Wrong Prediction"
    print(f"Predicted Value: {prediction}, Actual Value: {actual}, {is_prediction_correct}")
    results.append(1 if prediction==actual else 0) #for calculating accuracy, appending 1 if equal and 0 if not
    Y_pred.append(prediction)
```

```python
print(f"Accuracy of the Model: {(sum(results) * 100)/len(results)}")
```
